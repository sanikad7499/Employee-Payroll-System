 Employee Payroll & Workforce Analytics System Objective: Develop a payroll management system for a company using Core Python and NumPy. Modules: • Add employee details (ID, Name, Department, Salary). • Store and retrieve records using text files. • Create Employee class with OOP concepts. • Validate salary and inputs using exception handling. • Calculate salary statistics (mean, median, standard deviation, highest, lowest). • Apply increments using scalar operations and broadcasting. • Perform department-wise matrix operations. • Generate random employee datasets for testing. • Filter employees based on salary using Boolean indexing. Learning Outcomes: Students will build an intermediate payroll application demonstrating Python programming, data storage, OOP, exception handling and numerical analysis using NumPy.

```python
import numpy as np

# Employee Class
class Employee:
    def __init__(self, emp_id, name, department, salary):
        self.emp_id = emp_id
        self.name = name
        self.department = department
        self.salary = salary

    def display(self):
        print(self.emp_id, self.name, self.department, self.salary)

    def save(self):
        file = open("employees.txt", "a")
        file.write(f"{self.emp_id},{self.name},{self.department},{self.salary}\n")
        file.close()


employees = []


# Add Employee
def add_employee():
    try:
        emp_id = input("Enter Employee ID: ")
        name = input("Enter Name: ")
        department = input("Enter Department: ")

        salary = float(input("Enter Salary: "))

        if salary <= 0:
            raise ValueError

        emp = Employee(emp_id, name, department, salary)
        employees.append(emp)
        emp.save()

        print("Employee Added Successfully.")

    except ValueError:
        print("Invalid Salary!")


# Display Employees
def display_employees():
    if len(employees) == 0:
        print("No Employee Found")
    else:
        print("\nEmployee Details")
        for emp in employees:
            emp.display()


# Read File
def read_file():
    try:
        file = open("employees.txt", "r")
        print("\nEmployee Records From File")
        print(file.read())
        file.close()

    except FileNotFoundError:
        print("File Not Found")


# Salary Statistics
def salary_statistics():
    if len(employees) == 0:
        print("No Data")
        return

    salary = np.array([emp.salary for emp in employees])

    print("\nSalary Statistics")
    print("Mean Salary :", np.mean(salary))
    print("Median Salary :", np.median(salary))
    print("Standard Deviation :", np.std(salary))
    print("Highest Salary :", np.max(salary))
    print("Lowest Salary :", np.min(salary))


# Salary Increment
def increment_salary():
    if len(employees) == 0:
        print("No Data")
        return

    salary = np.array([emp.salary for emp in employees])

    increment = salary + 5000

    print("\nSalary After Increment")
    print(increment)


# Department Matrix
def department_matrix():
    if len(employees) == 0:
        print("No Data")
        return

    salary = np.array([emp.salary for emp in employees])

    matrix = salary.reshape(-1, 1)

    print("\nDepartment Salary Matrix")
    print(matrix)


# Random Employee Data
def random_dataset():
    random_salary = np.random.randint(25000, 90000, 10)

    print("\nRandom Salary Dataset")
    print(random_salary)


# Salary Filter
def filter_salary():
    if len(employees) == 0:
        print("No Data")
        return

    salary = np.array([emp.salary for emp in employees])

    limit = float(input("Show salaries greater than: "))

    print(salary[salary > limit])


# Menu
while True:

    print("\n====== Employee Payroll System ======")
    print("1. Add Employee")
    print("2. Display Employees")
    print("3. Read Employee File")
    print("4. Salary Statistics")
    print("5. Apply Salary Increment")
    print("6. Department Matrix")
    print("7. Random Dataset")
    print("8. Filter Salary")
    print("9. Exit")

    try:
        choice = int(input("Enter Choice: "))

        if choice == 1:
            add_employee()

        elif choice == 2:
            display_employees()

        elif choice == 3:
            read_file()

        elif choice == 4:
            salary_statistics()

        elif choice == 5:
            increment_salary()

        elif choice == 6:
            department_matrix()

        elif choice == 7:
            random_dataset()

        elif choice == 8:
            filter_salary()

        elif choice == 9:
            print("Thank You!")
            break

        else:
            print("Invalid Choice")

    except ValueError:
        print("Please Enter Correct Input")
```
